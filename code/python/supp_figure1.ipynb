# Supplementary material figure 1: Words and nonwords accuracy at each control parameter setting at the end of training

## Import libraries

```python
import pandas as pd
import altair as alt
from helper import RawData, control_space_heatmap
```

## Import part I raw data

```python
control_params = ["hidden_units",  "p_noise", "learning_rate"]

raw = RawData("../../data/data_part1_1250.csv")
df_word = raw.get(epoch_equal=1.0, word_type="word")
df_word = df_word.groupby(control_params).mean().reset_index()

df_nonword = raw.get(epoch_equal=1.0, word_type="nonword")
df_nonword = df_nonword.groupby(control_params).mean().reset_index()
```

## Plot end of training accuracy

```python
plot_word = control_space_heatmap(
    df_word,
    title="a. Word",
    var="score",
    color_scheme="redyellowgreen",
    domain=(0.6, 1),
    font_size=24,
    epsilon_label="Epsilon (panel column)"
)

plot_word
```

```python
plot_nonword = control_space_heatmap(
    df_nonword,
    title="b. Nonword",
    var="score",
    color_scheme="redyellowgreen",
    domain=(0.6, 1),
    font_size=24,
    epsilon_label=""
)

plot_nonword
```
